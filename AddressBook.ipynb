
# Project: Address Book<a href="#Project:-Address-Book" class="anchor-link">¶</a>

## Intro<a href="#Intro" class="anchor-link">¶</a>

In today's project we will do some practice with requests and regex by
looking through the Faculty and Staff Directory webpage of the School of
Pharmacy and getting contact information such as name, phone number,
email address, and office address.

## Get Imports<a href="#Get-Imports" class="anchor-link">¶</a>

First we put the imports re and requests which we need to get the url
and all of its contents in the url which we will convert into string
form.

In \[1\]:

    import re

In \[2\]:

    import requests

## URL and Regex Patterns<a href="#URL-and-Regex-Patterns" class="anchor-link">¶</a>

Down below we will get the url for the University at Buffalo's pharmacy
department's faculty and staff which we will represent as the variable
url. Next, we will get the contents of the page in string form and from
that string we will get the url for every body in the pharmacy faculty
and staff which can be found in the string html variable which we
represent as pharmacy. Regex is of great help here and we will get all
the url by using the regex pattern that we represented as the variable
pattern which will get the url from the \$\<a class="facultyindexname"\$
tag and we will stop the regex when the string gets into a double quote
which is represented as \$(\[^"\]+)\$ which is the regex synthax that
will get all the string up until the double quote. Atags variable
represents all the urls for all of the faculty and staff for the
pharmacy department which is represents as a list of strings and will
loop through the list in the next cell. Last but not least, the next
four variables represent the regex syntax that will look in an entire
string for patterns that contain the name, office address, phone, and
email in a string and in this case a html string.

In \[3\]:

    url = 'https://pharmacy.buffalo.edu/faculty-staff.html' #Url of University at Buffalo's pharmacy faculty and staff
    pharmacy  = requests.get(url).text #Html code for the url 
    pharmacy  = pharmacy.replace('>','>\n') #Html code in a more readable format
    pattern = r'<a class="facultyindexname" href="([^"]+)' #Regex pattern to get all the urls of the faculty and staff
    atags = re.findall(pattern,pharmacy) #All urls in a list of strings
    name_pattern = r'<span id="lblFacultyName">[\r\n]+([^\r\n][^<]+)' #Regex pattern to get all the names of the faculty/staff
    office_pattern = r'<span id="lblAcademicOfficeAddress">+[\r\n]+((?![<br/>])[0-9a-zA-Z, \r\n\t]+)'#Office regex pattern
    phone_pattern = r'Phone: ([^<]+)' #Regex pattern to get all the phone numbers of the faculty/staff
    email_pattern =  r'mailto:(\w+@buffalo.edu)' #Regex pattern to get all the emails of the faculty/staff  

## Contact Information<a href="#Contact-Information" class="anchor-link">¶</a>

In the cell below we will loop through the list of strings which we
represents as the variable atags and each string is the url for each of
the faculty and staff for the pharmacy department. Inside the for loop
we will get the contents of the current url we are investigating in the
for loop by using \$requests.get()\$ which has a string as the parameter
which represents the url and then do \$.text\$ to get the contents of
the html in string format and then use the \$replace()\$ method to put
in a readable format. This process will be represented as the variable
personInfo. After that, we will use \$re.findall()\$ method which takes
in two parameters: a pattern and the string that we are looking for to
see if it contains the pattern. The variables names, phones, offices,
and emails represent all the portions of the html string defined as
personInfo that contains the patterns defined in the cell above. Last
but not least, we will print the information gathered for each of the
urls.

In \[4\]:

    for ur in atags: #All urls
        personInfo = requests.get(ur).text #Html code for website of a specific faculty/staff member
        personInfo = personInfo.replace('>','>\n') #A more readable format of website of a specific faculty/staff member
        names = re.findall(name_pattern,personInfo) #Name of faculty/staff member
        phones = re.findall(phone_pattern,personInfo) #Phone number of faculty/staff member
        emails = re.findall(email_pattern,personInfo) #Email address of faculty/staff member
        offices = re.findall(office_pattern,personInfo) #Office address of faculty/staff member
        if len(names)>0:
            print(names[0])
            if len(offices)>0: print(offices[0])
            else: print('No Address')
            if len(phones)>1: print(phones[0])
            else: print('No Phone')
            if len(emails)>1: print(emails[0])
            else: print('No Email')
            print('-'*20)
