## Getting All Possible Words for a Text<a href="#Getting-All-Possible-Words-for-a-Text" class="anchor-link">¶</a>

First, we want to get all possible words for a text and in order to do
that we must read a file. After we read the file then we must split the
resulting string from reading the file by double spaces which we call by
the variable words. We know that a text has upper case and lower case
versions of a same word, but in Python those are considered different
strings so we want to convert all words into lower case words and we
will do that by looping through all the strings in a list and use the
\$lower()\$ method to convert any upper case strings into lower case and
then create a variable to keep track of this process. Then, we add that
lower case string into a list, however, we need to make sure that the
string we are adding to a list is an actual word and not just some
random string so will use the \$.isalnum\$ method which checks if all
characters in a string are alphanumeric meaning a string only contains
alphabet letters and numbers and if not then it returns False. Basically
we will add a string only if .isalnum is true and if this condition is
not true then we will not add this string into a list. Finally, after we
are done adding all the possible alphanumerical strings into a list then
we are done looping through the words list and then we will return the
new list with all the lower case words.

    def allWordsForFile(file):
        with open(file,'r',encoding='utf-8') as f: 
            s = f.read() #The string version of a file
            words = s.split() #List version of the String in the previous variable
            lowercasewords = [] #Initializing list of lower case words
            for s in words: #All strings
                word = s.lower() #Convert all characters in a string to lowercase
                lowercasewords.append(''.join(filter(str.isalnum, word))) #Only alphanumerical strings
        return lowercasewords #All Words

Now, we will set variables equal to the results of the method we defined
above and use those variables for further cells.

## Frequency of Words<a href="#Frequency-of-Words" class="anchor-link">¶</a>

Next in the process is for us to find the frequency of all the words in
a file or the amount of the times a word appears in a file. We will do
this by using a dictionary and in this dictionary the keys will
represents the words and the values will represent the amount of times
the words appear in a text file. In order to do this we will initialize
an empty dictionary and then loop through the list of all the words and
initially set all the values of the keys to 0. Then, we will loop
through the words again and increment a value for a key by one every
time it appears in a list. As a result, the frequency for each key is
determined and we finally got the dictionary with the words matching
with the frequency of the words.

In \[8\]:

    def frequencyOfWords(words):
        mydict = {} #Initializing dictionary
        for word in words: #Get all words in words list
            mydict[word] = 0 #Initially set each value to 0.
        for w in words: #Loop again
            mydict[w] += 1 #Every time word in words list matches key increment by 1
        return mydict #Dictionary with words and frequencies of words

Now we set variables equal to the results of the method above for word
lists from different files.

## Most Common Words for a Text File<a href="#Most-Common-Words-for-a-Text-File" class="anchor-link">¶</a>

Now we want to see the most common words for a text file. In order to do

third parameter which we will represent as reverse. What this
\$sorted()\$ method will do get all the key-value pairs in a dictionary
and sort it by the key and in this case key will be represented by
lambda x and lambda is an anonymous function which works like a regular
function expect we do not need a name and in this function we will sort

        freqList = [word[1] for word in common]
        wordds = [word[0] for word in common]

        plt.ylabel('Frequency') #y-axis label

        plt.title(titleforgraph) #Name of title

## Frequencies Based on Word Ranking<a href="#Frequencies-Based-on-Word-Ranking" class="anchor-link">¶</a>

Now we want to graph the frequencies of every word in the separate text
files not just the most common words based on their frequency ranking.
What this means is the most frequent word in a text file will be ranked
number 1 and the second most frequent will be ranked number 2 and so on
so far. This ranking will be represented on the x-axis and the frequency
for each ranking will be represented in the y-axis. We will do this
process in the method below and just like the \$PlotMostCommonWords()\$
function we will use list comprehensions and it will also take in a list
of all the words and their frequencies and the title of the graph, but
instead of doing a bar graph we will use the \$plt.loglog()\$ method
which will do a graph on a logarithmic scale. The reason we will be
doing this is because there are too many words in the text file and the
frequencies are in a large range of numbers that need to be displayed so
this method will allow us to show it graphically.

In \[27\]:

    def PlotWordRank(allOfThem,titleofPlot):
        freqList = [word[1] for word in allOfThem]
        rankings = [allOfThem.index(word) + 1 for word in allOfThem]
        plt.loglog(rankings,freqList) #x and y logarithmically
        plt.ylabel('Frequency') #y-axis label
        plt.xlabel('Ranking') #x-axis label
        plt.title(titleofPlot) #Name of title
        plt.xticks(rotation=30)
