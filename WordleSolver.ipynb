## Introduction<a href="#Introduction" class="anchor-link">¶</a>

Information Theory is the mathematical study of the amount of
information in data that could be transferred, stored, or retrieved
using universal mathematical units. One example of these mathematical
units are bits. Applications of Information Theory could be used for the
everyday life and one example of how Information Theory could be used is
in the solving of the popular word game Wordle. Words could be stored in
bits and Information Theory go hand to hand with this due to the fact
that the bits and the words associated with the bits could possible show
different patterns such as the number of possible words associated with
a guessed word that are ready to be unraveled and from this unraveling
Wordle can be solved more efficiently.

## Possible Words Method<a href="#Possible-Words-Method" class="anchor-link">¶</a>

Before the solving of the Wordle puzzle using Information Theory concept
like bits are in order, what must be done is to get all possible words
from a particular guess. What this method will do is have three
parameters which are a word that will be a guess, a parameter called
evaluation which will have a list of integers between 0 and 2. 0 means
that the letter in a guess does not appear in the actual word, 1 means
that the letter in a guess appears in the actual word but in different
positions, and 2 means that the letter in the guess appears in the same
position as the actual word. From that code that was given, a list of
tuples with each tuple having a number between 0 and 2 and a
corresponding letter. This will be done for when the positions are known
and when the positions are known. Next, a variable called possible_words
will just be a list of words and from there get every word in the words
list an find where they are removed or not removed and do necessary
conditionals to ensure all words that fit a criteria for a guess word
are in order. If the words match according to known letters and words
being removed over and over again due to the words not matching in any
specific criteria. Last but not least, the list of possible words from a
guess will be returned.

In \[4\]:

    def get_possible_words(guess_word,evaluation,words):
        known_letters_known_positions = [(i,letter) for i,letter in enumerate(guess_word) if evaluation[i] == 2]
        known_letters_unknown_positions = [(i,letter) for i,letter in enumerate(guess_word) if evaluation[i] == 1]
        known_missing_letters = [letter for i,letter in enumerate(guess_word) if evaluation[i] == 0]
        possible_words = list(words)
        for word in words:
            remaining_letters = list(word)
            removed = False
            for i,letter in known_letters_known_positions:
                if word[i] != letter:
                    possible_words.remove(word)
                    removed = True
                    break
                else:
                    remaining_letters.remove(letter)
            if not removed:
                for i,letter in known_letters_unknown_positions:
                    if not (letter in remaining_letters and word[i] != letter):
                        possible_words.remove(word)
                        removed = True
                        break
                    else:
                        remaining_letters.remove(letter)
            if not removed:
                for letter in known_missing_letters:
                    if letter in remaining_letters:
                        possible_words.remove(word)
                        removed = True
                        break
        return possible_words

## Solving Wordle Without Information Theory<a href="#Solving-Wordle-Without-Information-Theory" class="anchor-link">¶</a>

Although the main objective is to solve Wordle by applications of
Information Theory here is an another way of solving Wordle that will
have some elements used in the solving used by Information Theory way.
First, the solve_wordle() method will take in three parameters which are
the actual word, a guess, and the total amount of words in Wordle. Next,
a variable called evaluation will be used which will called the
evaluate_guess() method that was imported from the Wordle game which
will take in as parameters the actual word and a guessed word and output
a number between 0 and 2. What the number 0,1,2 represent was explained
in the previous method. Evaluation will be used a parameter to call an
another method: get_possible_words which will take in three parameters:
a guess word, evaluation, and the total number of words in Wordle and
then after that method is called it will be stored in a variable called
possible_words which works likes the previous method that was called and
shown in the print statement above. The variable guesses which
represents the number of guesses it takes to get a correct word and will
initially be 1 but it will be updated in the while loop that is coming
next. Inside the while loop, an another guess word will be gotten from
the possible_words variable and that guess word will be the first word
in the possible_words list. Then, the variable evaluation will be
updated for this new guess word and from the updated evaluation method
and updated possible words will be in order from the updated guess word
and evaluation and then the number of guesses will be incremented by 1.
What the while loop will do is keep doing this process over and over
again until the possible_words list is 1 meaning there is only one word
left which means that is the correct word. Lastly, the solve_wordle
method will return the correct word and the number of guesses it took to
get that correct word.

In \[6\]:

    def solve_wordle(correct_word,guess_word,words):
        evaluation = evaluate_guess(correct_word,guess_word)
        possible_words = get_possible_words(guess_word,evaluation,words)
        guesses = 1
        while len(possible_words) > 1:
            g_word = possible_words[0]
            evaluation = evaluate_guess(correct_word,g_word)
            possible_words = get_possible_words(g_word,evaluation,possible_words)
            guesses += 1
        return correct_word,guesses

## Solving Wordle Using Information Theory<a href="#Solving-Wordle-Using-Information-Theory" class="anchor-link">¶</a>

A bit can be found using the formula \$log_2(1/p(x))\$ in which p(x)
represents the probability of a pattern from a guess in Wordle. The
information_values_of_words() method will take in the actual word, a
guess, and all the words in Wordle. ProbabilityList will store all the
probabilities of all the patterns from all the possible words that will
be used as guesses. To get the probability the length of the possible
words list will be divided by the number of words in Wordle. Then, like
the method before a while loop will be use with the same process as
before but instead each possible word in bit form will be added to the
information list and then after the while loop the information list
which will contain all the possible words and the guess will be
returned.

    import math

    def information_values_of_words(correct_word,guess_word,words):
        probabilityList = []
        InfoList = []
        eva = evaluate_guess(correct_word,guess_word)
        possible_words = get_possible_words(guess_word,eva,words)
        I = -math.log2(len(possible_words) / len(words))
        InfoList.append(I)
        while len(possible_words) > 1:
            g_word = possible_words[0]
            eva = evaluate_guess(correct_word,g_word)
            possible_words = get_possible_words(g_word,eva,possible_words)
            p = len(possible_words) / len(words)
            I = math.log2(1/p)
            InfoList.append(I)
        return InfoList
